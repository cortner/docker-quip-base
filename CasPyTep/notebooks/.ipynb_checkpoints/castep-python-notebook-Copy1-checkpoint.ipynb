## Atomic Simulation Environment (ASE)

- Within atomistic modelling, standard for scripting interfaces is ASE
- Wide range of calculators, flexible (but not too flexible) data model for Atoms objects
- Can use many codes as drop-in replacements:

<img src="ase-calculators.png" width="80%">

- Collection of parsers aids validation and verification (cf. DFT $\Delta$-code project)
- Coupling $N$ codes requires maintaining $N$ parsers/interfaces to be maintained, not $N^2$ input-output converters
- High-level functionality can be coded generically, or imported from other packages (e.g. `spglib`, `phonopy`) using minimal ASE-compatible API

https://wiki.fysik.dtu.dk/ase/

# File-based interfaces vs. Native interfaces

- ASE provides file-based interfaces to a number of codes - useful for high throughput
- However, file-based interfaces to electronic structure codes can be slow and/or incomplete and parsers are hard to keep up to date and robust
- Standardised output (e.g. chemical markup language, XML) and robust parsers are part of solution
- [NoMaD Centre of Excellence](http://nomad-coe.eu/) will produce parsers for top ~40 atomistic codes
- Alternative: native interfaces provide a much deeper wrapping, exposing full public API of code to script writers
    - e.g. GPAW, LAMMPSlib, this new CasPyTep Castep Python interface
- Future proofing: anything accessible from Python is available from other high-level languages too (e.g. Julia)

   
# `f90wrap` adds derived type support to `f2py`

- There are many automatic interface generators for C++ codes (e.g. SWIG or Boost.Python), but not many support modern Fortran
    - [f2py](https://sysbio.ioc.ee/projects/f2py2e/) scans Fortran 77/90/95 codes, generates Python interfaces for individual routines
    - No support for modern Fortran features: derived types, overloaded interfaces
- Despite recent increases in use of high-level languages, there's still lots of high-quality Fortran code around
- Number of follow up projects, none had all features we needed
- `f90wrap` addresses this by generating an additional layer of wrappers, adding support for derived types, module data, efficient array access, Python 2.6+ and 3.x

https://github.com/jameskermode/f90wrap    

# Example: wrapping  the `bader` code

- Widely used code for post-processing charge densities to construct Bader volumes
- Python interface would allow code to be used as part of workflows without needing file format converters, I/O, etc.
- Downloaded [source](http://theory.cm.utexas.edu/henkelman/code/bader/), used `f90wrap` to *automatically* generate a deep Python interface with very little manual work

Generation and compilation of wrappers:

    f90wrap -v -k kind_map -I init.py -m bader \
         kind_mod.f90 matrix_mod.f90 \
         ions_mod.f90 options_mod.f90 charge_mod.f90 \
         chgcar_mod.f90 cube_mod.f90 io_mod.f90 \
         bader_mod.f90 voronoi_mod.f90 multipole_mod.f90

    f2py-f90wrap -c -m _bader f90wrap_*.f90 -L. -lbader

````python
# collect Bader volumes associated with atom #5
mask = np.zeros_like(rho, dtype=bool)
for v in (bdr.nnion == atom+1).nonzero()[0]:
    mask[bdr.volnum == v+1] = True
    
plot(si.positions[:, 0], si.positions[:, 1], 'k.', ms=20)
plot(si.positions[atom, 0], si.positions[5, 1], 'g.', ms=20)
imshow(rho[:,:,0],  extent=[0, si.cell[0,0], 0, si.cell[1,1]])
imshow(mask[:,:,0], extent=[0, si.cell[0,0], 0, si.cell[1,1]], alpha=.6)
````

# Wrapping Castep with f90wrap - CasPyTep

- `f90wrap` can now wrap large and complex codes like Castep - providing deep access to internal data on-the-fly 
- Summer project in 2014 by Greg Corbett at STFC built proof-of-principle Castep/Python interface
- Results described in [RAL technical report](https://epubs.stfc.ac.uk/work/18048381)
- Took this a bit further at Castep codefest in 2015
- Warwick MSc student Sebastian Potthoff wrote his dissertation on CasPyTep in 2016, adding MPI support and looking at performance of nudged elastic band 
- Since then I've tidied it up a little, fixed a few bugs, and added a minimal high-level ASE-compatibility layer, but there's lots still to be done

![Castep](http://www.castep.org/files/CASTEP_Logo_mini-01.png)

# CasPyTep Current Features

- Set of source files currently wrapped is as follows, can be easily expanded:

        Utility:     constants.F90 algor.F90 
                     comms.serial.F90
                     io.F90 trace.F90 
                     license.F90 buildinfo.f90
        Fundamental: parameters.F90 cell.F90 basis.F90 
                     ion.F90 density.F90 wave.F90
        Functional:  model.F90 electronic.F90 firstd.f90
        
- Already far too much to wrap by hand: 
   - 35 kLOC Fortran and 55 kLOC Python auto-generated 
   - 23 derived types 
   - ~2600 subroutines/functions

# What is wrapped?
- Module-level variables: `current_cell`, etc.
- Fortran derived types visible as Python Classes: e.g; `unit_cell`, `model_state`
- Arrays (including arrays of derived types) - no copying necessary to access data in numerical arrays e.g. `current_cell%ionic_positions`
- Documentation strings extracted from source code
- Dynamic introspection of data and objects
- `io_abort()` raises `RuntimeError` exception
- Minimal ASE-compatible high level interface

## Single point calculation

This uses the ASE-compatible interface provided by the `CastepCalculator` class.

````python
from ase.lattice.cubic import Diamond
atoms = Diamond('Si')
calc = castep.calculator.CastepCalculator(atoms=atoms, calculate_stress=True, fixed_npw=True)
atoms.set_calculator(calc)
print 'energy'
e = atoms.get_potential_energy()
print e
print 'force'
f = atoms.get_forces()
print f
print 'stress'
s = atoms.get_stress()
print s
````

## Postprocessing

````python
%pylab inline
cell = calc.model.cell

# plot the ions with black balls
p = cell.ionic_positions.copy()
p = p[:, :, 0] # first species only
# absolute positions of ions in Ang
from ase.units import Bohr
p = np.dot(cell.real_lattice.T, p)
xi, yi, zi = p*Bohr
scatter(xi, yi, s=100, c='k')

# plot the charge density
den = calc.model.den.real_charge.copy()
ngx, ngy, ngz = (castep.basis.get_ngx(), 
                 castep.basis.get_ngy(), 
                 castep.basis.get_ngz())
den3 = (den.reshape((ngx, ngy, ngz), order='F') / 
        castep.basis.get_total_fine_grid_points())
imshow(den3[:, :, ngz/2], 
       extent=[0, atoms.cell[0,0], 
               0, atoms.cell[1,1]])
````

So far this is just analysis/post-processing, but could easily go beyond this and **steer calculations** based on results of e.g. Bader analysis.

## Updating data inside a running Castep instance

We can move the ions and continue the calculation without having to restart electronic minimisation from scratch (or do any I/O of `.check` files etc.).

Here's how the core of the `CastepCalculator.calculate()` method is coded in Python:

```python
if 'energy' not in self.results:
    new_cell = atoms_to_cell(atoms, kpts=self.kpts)
    castep.cell.copy(new_cell, self.current_cell)
    self.model.wvfn.have_beta_phi = False
    castep.wave.wave_sorthonormalise(self.model.wvfn)
    self.model.total_energy, self.model.converged = \
    electronic_minimisation(self.model.wvfn, 
                            self.model.den, 
                            self.model.occ, 
                            self.model.eigenvalues, 
                            self.model.fermi_energy)
    self.results['energy'] = io_atomic_to_unit(self.model.total_energy, 'eV')
```

```python
if 'forces' not in self.results:
    forces = np.zeros((3, self.current_cell.num_ions), order='F')
    castep.firstd.firstd_calculate_forces(self.model.wvfn, 
                                          self.model.den, 
                                          self.model.occ, 
                                          self.model.eigenvalues, 
                                          forces)
    self.results['forces'] = forces.T
    self.results['forces'][...] *= EV_PER_ANG
```

Allows embedded Castep to be **efficiently** used as a standard ASE calculator, with all of the existing high-level algorithms: geometry optimisation, NEB, basin hopping, etc.
 - Compared to file-based interface, save overhead of starting Castep each time
 - Reuse electronic model from one ionic configuration to the next
 - Wavefunction and charge density extrapolation possible just as in MD

````python
from ase.optimize import LBFGS
atoms.rattle(0.01)
opt = LBFGS(atoms)
opt.run(fmax=0.1)
````

## Testing new algorithms with Castep

Having a Python interface also makes it quick to try out new high-level algorithms or connect things in new ways. 

````python
from ase.lattice import bulk
from ase.optimize import LBFGS

import castep

atoms = bulk('Si', cubic=True)
atoms *= (2, 1, 1)

# compress system along x, long-range deformation hard
# for standard minimisers
s = atoms.get_scaled_positions()
s[:, 0] *= 0.98
atoms.set_scaled_positions(s)
calc = castep.calculator.CastepCalculator(atoms=atoms)
atoms.set_calculator(calc)    
opt = LBFGS(atoms)
opt.run(fmax=0.05)
````

##  Convergence testing

This is an example of using the native Castep interface directly rather than the ASE compatibility layer. We increase the plane wave cutoff energy in steps of 10% until energy changes by less than $10^{-3}$ Ha. (Yes, I know this isn't the best way to do convergence testing...)

````python
energy_tol = 1e-3
current_params = castep.parameters.get_current_params()
current_params.cut_off_energy = 7.0

cutoffs = []
energy = []

while True:
    castep.basis.basis_initialise(current_params.cut_off_energy)
    current_params.fine_gmax = (current_params.fine_grid_scale *
                                np.sqrt(2.0*current_params.cut_off_energy))
    castep.ion.ion_real_initialise()
    model = castep.model.Model_State()
    (model.total_energy, model.converged) = \
     castep.electronic.electronic_minimisation(model.wvfn, 
                                               model.den, 
                                               model.occ,
                                               model.eigenvalues, 
                                               model.fermi_energy)
    current_params.cut_off_energy *= 1.1

    print 'cutoff %.2f energy %.5f' % (current_params.cut_off_energy, 
                                       model.total_energy)
    cutoffs.append(current_params.cut_off_energy)
    energy.append(model.total_energy)
    
    if len(energy) > 2 and abs(energy[-1] - energy[-2]) < energy_tol:
        print 'converged at cutoff', cutoffs[-1]
        break
````

````python
plot(cutoffs, energy, 'o-')
````
